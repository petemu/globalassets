```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)", color="orange", linewidth=3)
plt.title("My First Notebook Plot")
plt.xlabel("X axis")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```
